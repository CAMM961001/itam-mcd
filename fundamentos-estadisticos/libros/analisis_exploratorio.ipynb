```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
path_universal = "C:\\Users\\Miguel\\Documents\\Github\\itam-mcd"
os.listdir(path_universal)
```

```python
path_asignatura = "fundamentos-estadisticos"
path_asignatura = os.path.join(path_universal, path_asignatura)
os.listdir(path_asignatura)
```

```python
#Se obtiene dataset original
file_path = "datos\\propinas.csv"
propinas = pd.read_csv(os.path.join(path_asignatura, file_path))
propinas.head(10)
```

```python
#Se genera tabla orden_cuenta
orden_cuenta = pd.DataFrame(propinas.sort_values(by='cuenta_total')['cuenta_total'])
orden_cuenta.reset_index(drop=True, inplace=True)
orden_cuenta['orden_cuenta'] = pd.Series(np.arange(1,orden_cuenta.shape[0]+1))
orden_cuenta['f'] = (orden_cuenta['orden_cuenta'] - 0.5)/len(orden_cuenta)

orden_cuenta
```

```python
#Se hacen las gráficas de cuantiles
x1_feat, x2_feat, y_feat = "orden_cuenta", "f", "cuenta_total"

#Tamaño de la figura
plt.figure(figsize=(14,6))

#Graficando para X1
plt.subplot(121)
plt.scatter(x=orden_cuenta[x1_feat],
           y=orden_cuenta[y_feat],
           c='tab:red',
           alpha=0.5)

#Styling
plt.title("Cuenta total", fontsize=16, loc='left')
plt.xlabel(x1_feat, fontsize=14)
plt.ylabel(y_feat, fontsize=14)
plt.grid(alpha=0.25)

#Graficando para X2
plt.subplot(122)
plt.scatter(x=orden_cuenta[x2_feat],
           y=orden_cuenta[y_feat],
           c='tab:red',
           alpha=0.5)
plt.plot(orden_cuenta[x2_feat],
        orden_cuenta[y_feat],
        c='black',
        linewidth=0.5)

#Styling
plt.title("Cuantiles de cuenta total", fontsize=16, loc='left')
plt.xlabel(x2_feat, fontsize=14)
plt.ylabel(y_feat, fontsize=14)
plt.grid(alpha=0.25)

plt.show()
```
