```python
casas_f = prices[features.values()]
casas_f.set_axis(list(features.keys()), axis=1, inplace=True)
casas_f.info()
```

Se hace un **join** por la derecha entre `casas_f` y `geom`.

Aparecerán las variables `42: lat` y `43: long` dentro del dataset de casas

```python
casas_f = pd.merge(left=casas_f,
        right=geom,
        on='nombre_zona')

casas_f.info()
```

Se crean **variables por $m^2$ y $K USD**
* area_sotano_m2
* area_1er_piso_m2
* area_2o_piso_m2
* area_hab_m2
* area_garage_m2
* area_lote_m2
* precio_miles
* valor_misc_miles
* precio_m2_miles

```python
m2 = 0.092903

#Variables por metro cuadrado
casas_f['area_sotano_m2'] = casas_f['area_sotano'] * m2
casas_f['area_1er_piso_m2'] = casas_f['area_1er_piso'] * m2
casas_f['area_2o_piso_m2'] = casas_f['area_2o_piso'] * m2
casas_f['area_hab_m2'] = casas_f['area_habitable_sup'] * m2
casas_f['area_garage_m2'] = casas_f['area_garage'] * m2
casas_f['area_lote_m2'] = casas_f['area_lote'] * m2

#Variables por miles de dolares
casas_f['precio_miles'] = casas_f['precio'] / 1_000
casas_f['valor_misc_miles'] = casas_f['valor_misc'] / 1_000
casas_f['precio_m2_miles'] = casas_f['precio_miles'] / casas_f['area_hab_m2']

casas_f.info()
```

Se descartan variables que no serán utilizadas

```python
cols2drop = ['area_sotano',
             'area_1er_piso',
             'area_2o_piso',
             'area_habitable_sup',
             'area_garage',
             'area_lote',
             'precio',
             'valor_misc']

casas = casas_f.drop(columns=cols2drop)
casas.info()
```
