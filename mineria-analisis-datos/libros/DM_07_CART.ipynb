```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

```python
# Load iris data
iris = load_iris()
X = iris.data
y = iris.target

print(X.shape)
print(y.shape)
```

```python
# Split training and test sets
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.1)
```

# Trees

```python
# Create and train a decision tree classifier
dt_model = DecisionTreeClassifier()
dt_model.fit(x_train, y_train)
```

```python
# print its performance on the training and test sets
print(f"Performance on the training set: {dt_model.score(x_train, y_train):.3f}")
print(f"Performance on the test set: {dt_model.score(x_test, y_test):.3f}")

# Hay un poco de sobre ajuste
```

# Random forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
# Create and train a random forest classifier
rf_model = RandomForestClassifier(n_estimators=30, max_depth=2)
rf_model.fit(x_train, y_train)
```

```python
# print its performance on the training and test sets
print(f"Performance on the training set: {rf_model.score(x_train, y_train):.3f}")
print(f"Performance on the test set: {rf_model.score(x_test, y_test):.3f}")

# Hay un poco de sobre ajuste
```
